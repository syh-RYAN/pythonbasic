##### if문
- if문의 바디는 나중에 작성할 예정이라면 pass 키워드를 통해 아무것도 하지 않고 오류 없이 코드를 작성 할 수 있다.
- if 조건식 :
    pass # TOBE
- if문의 조건식이 True일 때 실행되는 범위는 들여쓰기가 끝날 때까지이다. 반대로 if문의 조건식이 False일 때 들여쓰기가 끝날 때까지의 코드는 실행되지 않습니다.
- if문을 중첩해서 사용할 수 있습니다. 이때 중첩될수 있도록 들여쓰기를 한 번 더 해준다.

```python
age = int(input("나이를 입력해주세요"))

if age >=10:
    print("10살 이상입니다.")
    if age >=20:
        print("20살 이상입니다.")
        if age >=30:
            print("30살 이상입니다.")
            if age >=40:
                print("40살 이상입니다.")
            print("age >=30 구간입니다.")
        print("age >=20 구간입니다.")
    print("age >=10 구간입니다.")

print("늘 출력되는 구간입니다.")
```

##### range()함수
- range는 연속되는 숫자 요소들을 만들 때 활용
- range()함수는 슬라이싱과 **구조가 비슷**합니다. 다만, range()의 경우 함수의 형태로 쓰기 때문에 소괄호 안에 구분자는 당연히 콤마(,)입니다.
- 시작 인덱스는 포함(이상), 끝 인덱스는 불포함(미만)입니다. 시작 인덱스(default = 0)와 증감크기 (default = 1)는 생략 가능합니다.
- range()함수를 숫자들의 시퀀스로 이터레이트할 때 사용하면 편리
range변수 = range(시작,끝,증감크기)
시퀀스자료형변수 = range(시작:끝:증감크기)
##### range()함수로 list 생성하기
- 연속되는 정수를 생성할 때 range() 내장 함수를 사용하면 편리
- range()로 반환받은 결과 자료형은 range입니다. 이는 list()내장 함수를 활용해서 리스트로 변경 가능
ex) even_numbers = list(range(2,100,2)) -> 100미만의 2의 배수를 담은 리스트를 생성하는 방법

```python
even_numbers = list(range(2,100,2))
print(even_numbers)
```

##### 시퀀스 객체로 반복하기
- for문에 range() 대신 시퀀스 객체를 넣어서 사용할 수 있습니다.
- 시퀀스 자료형을 사용하여 for문을 looping하면 요소마다 순서에 맞게 꺼내와서 반복문을 실행합니다.

```python
for _ in("파","도","솔"):
    print(_,end=" ")
```

##### 시퀀스 객체로 뒤집어서 반복하기
- reversed()함수를 통해서 역순으로 변경 가능.
- 리서드 메서드에 있는 reverse()와는 다름 함수
- reversed(시퀀스) 함수는 반환값을 주지만 해당 시퀀스를 변경하지 않습니다.
- 리스트.reverse()함수는 리스트 요소 순서를 뒤집어주며 반환값이 없습니다.(None).
- 변수[::-1]

```python
for _ in reversed("일이삼사오육칠팔구십"):
    print(_,end=" ")
```

```python
for _ in(10,-42,25,103,21,77,24):
    print(_*10,end=" ")
```
