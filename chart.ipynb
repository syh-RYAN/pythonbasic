```python
# tight_layout() 명령을 실행하면 플롯 간의 간격을 자동으로 맞춰줌
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r-')
plt.title('time (s) ')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
# 컨투어 플롯
# 자료를 시각화하는 방법은 명암인 아닌 등고선

def f(x,y):
    return (1-x / 2 + x ** 5 + y **3) * np.exp(-x ** 2 -y **2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
XX,YY=np.meshgrid(x,y)
zz = f(XX,YY)

plt.title("Contour plots")
plt.contourf(XX,YY,zz,alpha=.75, cmap = 'jet')
plt.contour(XX,YY,zz,colors='black')
plt.show()
```

```python
# 3D 서피스 플롯
# Plot_wireframe(), plot_surface()명령 사용

from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```

```python
# 카운트 플롯 유니크한 값들이 몇개씩 있는지 시각적으로 확인
df = sns.load_dataset("titanic")
sns.countplot(x=df["class"])
```

```python
# hue 키워드 인수를 활용하여 분류할 두 번째 변수를 설정할 수 있다.
# data 키워드 인수에 DataFrame을 전달하고 x 키워드 인수에 DataFrame 중 매칭하고 시픈 column의 name값을 전달
sns.countplot(data=df, x="class", hue="alive")
```

```python
# 플롯을 수평하게
# y키워드 인수에 data로 넘긴 DataFrame의 column name 값을 전달
sns.countplot(data=df, y="deck", hue="alive")
```
